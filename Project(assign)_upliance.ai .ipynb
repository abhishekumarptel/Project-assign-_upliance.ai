```python
import pandas as pd

# Load the Excel file
file_path = 'Data Analyst Intern Assignment - Excel.xlsx'

# Read all sheets into a dictionary of DataFrames
all_sheets = pd.read_excel(file_path, sheet_name=None)

# Print the sheet names (keys of the dictionary)
print("Sheet names:", all_sheets.keys())

# Access data from each worksheet
sheet1 = all_sheets['UserDetails.csv']
sheet2 = all_sheets['CookingSessions.csv']
sheet3 = all_sheets['OrderDetails.csv']


# Print the content of one sheet
print(sheet1.head())  # Display the first few rows of the first sheet
```

```python
df
```

```python
sheet3['Rating'] = sheet3['Rating'].fillna(sheet3['Rating'].mode() [0])
```

```python
sheet3
```

```python
# Get unique rows from the entire DataFrame
sheet4 = sheet3.drop_duplicates('User ID')

print(sheet4)
```

```python
sheet5 = sheet2.drop_duplicates('User ID')

print(sheet5)
```

```python
# First merge sheet1 and sheet4
merged_df1 = pd.merge(sheet1, sheet4, on='User ID')

# Then merge the result with sheet5
merged_df = pd.merge(merged_df1, sheet5, on='User ID')

# Print the merged DataFrame
print(merged_df)
```

```python
merged_df
```

```python
# Analyzing the Relationship Between Cooking Sessions and User Orders
# Count orders per session
orders_per_session = merged_df.groupby('Session ID_x')['Order ID'].count().reset_index()
orders_per_session.rename(columns={'Order ID': 'Number of Orders'}, inplace=True)

# Show the number of orders for each session
print(orders_per_session)
```

```python
# Meal Type Popularity
# Count orders by meal type
orders_by_meal_type = merged_df.groupby('Meal Type_y')['Order ID'].count().reset_index()
orders_by_meal_type.rename(columns={'Order ID': 'Number of Orders'}, inplace=True)

# Sort by the number of orders
orders_by_meal_type = orders_by_meal_type.sort_values(by='Number of Orders', ascending=False)

# Show the result
print(orders_by_meal_type)
```

```python
# Identifying Popular Dishes
# Count orders by dish name
popular_dishes = merged_df.groupby('Dish Name_y')['Order ID'].count().reset_index()
popular_dishes.rename(columns={'Order ID': 'Number of Orders'}, inplace=True)

# Sort by the number of orders
popular_dishes = popular_dishes.sort_values(by='Number of Orders', ascending=False)

# Show the result
print(popular_dishes)
```

```python
#Exploring Demographic Factors That Influence User Behavior
# Create age groups
bins = [0, 18, 25, 35, 45, 55, 100]
labels = ['<18', '18-25', '26-35', '36-45', '46-55', '55+']
merged_df['Age Group'] = pd.cut(merged_df['Age'], bins=bins, labels=labels, right=False)

# Count orders by age group
orders_by_age = merged_df.groupby('Age Group')['Order ID'].count().reset_index()
orders_by_age.rename(columns={'Order ID': 'Number of Orders'}, inplace=True)

# Show the result
print(orders_by_age)
```

```python
# Orders by Location
# Count orders by location
orders_by_location = merged_df.groupby('Location')['Order ID'].count().reset_index()
orders_by_location.rename(columns={'Order ID': 'Number of Orders'}, inplace=True)

# Show the result
print(orders_by_location)
```

```python
#Favorite Meal and Number of Orders
# Count orders by favorite meal
orders_by_favorite_meal = merged_df.groupby('Favorite Meal')['Order ID'].count().reset_index()
orders_by_favorite_meal.rename(columns={'Order ID': 'Number of Orders'}, inplace=True)

# Show the result
print(orders_by_favorite_meal)
```

```python
#Create Visualizations
#1. Number of Orders by Meal Type
import matplotlib.pyplot as plt
import seaborn as sns
# Create a list of colors for each bar (e.g., a different color for each meal)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Customize the colors

# Assuming orders_by_meal_type is a DataFrame with meal type data
plt.figure(figsize=(8, 6))
sns.barplot(x='Meal Type_y', y='Number of Orders', data=orders_by_meal_type,palette=colors)
plt.title('Number of Orders by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
#2.Top 10 Popular Dishes
# Plot top 10 popular dishes
# Create a list of colors for each bar (e.g., a different color for each meal)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Customize the colors
plt.figure(figsize=(10, 6))
sns.barplot(x='Dish Name_y', y='Number of Orders', data=popular_dishes.head(10),palette=colors)  
plt.title('Top 10 Popular Dishes')
plt.xlabel('Dish Name')
plt.ylabel('Number of Orders')
plt.xticks(rotation=90)
plt.show()
```

```python
#3.Orders by Age Group
# Create a list of colors for each bar (e.g., a different color for each meal)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Customize the colors
# Plot orders by age group
plt.figure(figsize=(8, 6))
sns.barplot(x='Age Group', y='Number of Orders', data=orders_by_age, palette=colors)
plt.title('Orders by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Orders')
plt.show()
```

```python
#4.Orders by Location
# Create a list of colors for each bar (e.g., a different color for each meal)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Customize the colors
# Plot orders by location
plt.figure(figsize=(10, 6))
sns.barplot(x='Location', y='Number of Orders', data=orders_by_location, palette=colors)
plt.title('Orders by Location')
plt.xlabel('Location')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
#5.Orders by Favorite Meal
# Create a list of colors for each bar (e.g., a different color for each meal)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Customize the colors

# Plot orders by favorite meal
plt.figure(figsize=(10, 6))
sns.barplot(x='Favorite Meal', y='Number of Orders', data=orders_by_favorite_meal, palette=colors)
plt.title('Orders by Favorite Meal')
plt.xlabel('Favorite Meal')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```
